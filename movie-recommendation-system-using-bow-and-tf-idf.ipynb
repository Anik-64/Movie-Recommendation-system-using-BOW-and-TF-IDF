```python
movies = pd.read_csv('../input/movierecommendationsystem/tmdb_5000_movies.csv/tmdb_5000_movies.csv')
credits = pd.read_csv('../input/movierecommendationsystem/tmdb_5000_credits.csv/tmdb_5000_credits.csv')
```

```python
# merge data on title
movies = movies.merge(credits,on='title')
```

```python
movies.head(2)
```

```python
# For recommendation 
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
# this function helps to find the name filed from genres,keywords and convert it into a list
import ast
def convert(obj):
    genres_list = []
    for i in ast.literal_eval(obj):
        genres_list.append(i['name'])
    return genres_list
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
# this function helps to find the top 3 actors name and add it to a list
def cast_convert(obj):
    cast_list = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            cast_list.append(i['name'])
            counter += 1
        else:
            break
    return cast_list
```

```python
movies['cast'] = movies['cast'].apply(cast_convert)
```

```python
movies.head(2)
```

```python
movies.iloc[0].crew
```

```python
# this helps to find the director of a movie
def crew_convert(obj):
    crew_list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            crew_list.append(i['name'])
            break
    return crew_list
```

```python
movies['crew'] = movies['crew'].apply(crew_convert)
```

```python
movies.head(2)
```

```python
# to remove space between words
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(3)
```

```python
# this convert the overview string to list for concate with genres,keywords,cast and crew
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(3)
```

```python
# create new column tags 
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# new data frame
new_df_movies = movies[['movie_id','title','tags']]
```

```python
new_df_movies.head(3)
```

```python
# convert the tags list into a string
new_df_movies['tags'] = new_df_movies['tags'].apply(lambda x:" ".join(x))
```

```python
# convert the tags string into lowercase
new_df_movies['tags']=new_df_movies['tags'].apply(lambda x:x.lower())
```

```python
new_df_movies.head(3)
```

```python
apply_bow = new_df_movies
apply_tfidf = new_df_movies
```

```python
# to find the root word]
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
apply_bow['tags'] = apply_bow['tags'].apply(stem)
```

```python
apply_bow.iloc[0].tags
```

```python
# now apply BoW
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(apply_bow['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
vectors.shape
```

```python
# now apply cosine similarity 
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = apply_bow[apply_bow['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    
    for movies in movies_list:
        print(apply_bow.iloc[movies[0]].title)
```

```python
recommend('Avatar')
```

```python
# apply Tf-Idf
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(max_features=5000,stop_words='english')
```

```python
vectors_tfidf = tfv.fit_transform(apply_tfidf['tags'])
```

```python
vectors_tfidf.shape
```

```python
tfv.get_feature_names()
```

```python
# now apply cosine similarity 
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(vectors_tfidf,vectors_tfidf)
```

```python
sig
```

```python
sig[0]
```

```python
def recommend_tfidf(movie):
    movie_index = apply_tfidf[apply_tfidf['title'] == movie].index[0]
    distances = sig[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    
    for movies in movies_list:
        print(apply_bow.iloc[movies[0]].title)
```

```python
recommend_tfidf('Avatar')
```
